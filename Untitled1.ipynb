```python
#Initialization of basic data containers, gene annotation data, growth rates and cell volumes and selection of conditions to analyze.
import csv
from pandas.io.parsers import read_csv
import pandas as pd
from scipy.stats import gaussian_kde,linregress
import scipy.odr as od

pylab.rcParams['figure.figsize']=(10.0,7.0)

def get_cond(x):
    return x.split('_')[-1].split(';')[0]

def get_gene(x):
    res = x.split('__')
    if len(res)>1:
	return res[1]
    return 'missing'

# first load the data from the file.
cerevisae_data = read_csv('cerevisae_data.tsv',header=0,sep='\t')

# extract the growth condition and gene
cerevisae_data['cond'] = cerevisae_data['DATE_GR_PA_PAnewWindow;[2010,1,31,15,8,11]'].apply(get_cond)
cerevisae_data['gene'] = cerevisae_data['DATE_GR_PA_PAnewWindow;[2010,1,31,15,8,11]'].apply(get_gene)

#drop all irrelevant data
cerevisae_data = cerevisae_data[cerevisae_data['gene'] != 'missing']
cerevisae_data['GR_secs'] = cerevisae_data['GR_secs'].astype('float64')
cerevisae_data = cerevisae_data[cerevisae_data['PA_newWindow'] != '#NAME?']
cerevisae_data['PA_newWindow'] = cerevisae_data['PA_newWindow'].astype('float64')
cerevisae_data = cerevisae_data.dropna()
cerevisae_data = cerevisae_data[cerevisae_data['GR_secs']>0]
cerevisae_data = cerevisae_data[cerevisae_data['PA_newWindow']>0]
cerevisae_data = cerevisae_data[cerevisae_data['cond']!='NaCl']
cerevisae_data = cerevisae_data[cerevisae_data['cond']!='HS']
cerevisae_data = cerevisae_data[cerevisae_data.gene.str.startswith('Y')]

#extract average growth rates
groupedgr = cerevisae_data.groupby('cond')
gr = groupedgr['GR_secs'].mean()
gr = gr/3600
gr = 1/gr
gr = sort(gr)
cond_list = list(gr.keys().values)
cerevisae_prots = cerevisae_prots[cond_list]

#calculate protein fractions
#cerevisae_data['prot_conc']= cerevisae_data['GR_secs'] * cerevisae_data['PA_newWindow']
cerevisae_data['prot_conc']= cerevisae_data['GR_secs']
for i,row in cerevisae_data.iterrows():
    cerevisae_data.loc[i,'prot_conc']= cerevisae_data.loc[i,'PA_newWindow'] * gr[cerevisae_data.loc[i,'cond']]

grouped = cerevisae_data.groupby(['gene','cond'])
cerevisae_prots = grouped['prot_conc'].mean().unstack()
cerevisae_prots = cerevisae_prots.dropna()
cerevisae_prots = cerevisae_prots[cond_list]
cerevisae_prots = cerevisae_prots / cerevisae_prots.sum()
print cerevisae_prots
print gr
```
